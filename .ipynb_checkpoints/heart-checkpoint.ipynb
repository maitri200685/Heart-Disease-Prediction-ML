```python
df_encode = pd.get_dummies(df,drop_first=True)
```

```python
df_encode = df_encode.astype(int)
```

```python
df_encode.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['Age', 'RestingBP' , 'Cholesterol' , 'MaxHR','Oldpeak']
scaler = StandardScaler()
df_encode[numerical_cols] = scaler.fit_transform(df_encode[numerical_cols])
df_encode
```

```python
corr = df_encode.corr()

corr['HeartDisease'].sort_values(
    ascending=False
)
```

```python
from scipy.stats import chi2_contingency
import pandas as pd

# Significance level
alpha = 0.05

# Categorical columns
cat_features = ['Sex_M', 'resting BP', 'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA', 'FastingBS','RestingECG_Normal','RestingECG_ST', 'ExerciseAngina_Y','ST_Slope_Flat','ST_Slope_Up']

# Dictionary to store results
chi2_results = {}

# Chi-Square Test
for col in cat_features:

    contingency = pd.crosstab(
        df_encode[col],
        df_encode['HeartDisease']
    )

    chi2_stat, p_val, _, _ = chi2_contingency(contingency)

    decision = (
        'Reject Null (Keep Feature)'
        if p_val < alpha
        else 'Accept Null (Drop Feature)'
    )

    chi2_results[col] = {
        'Chi2 Statistic': chi2_stat,
        'P Value': p_val,
        'Decision': decision
    }

# Create DataFrame
chi2_df = pd.DataFrame(chi2_results).T

# Sort by p-value
chi2_df = chi2_df.sort_values(by='P Value')


chi2_df
```
